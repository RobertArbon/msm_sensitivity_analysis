```python
# ts = -tau/np.log(ev)
# -ts/tau = 1/log(ev)
# -tau/ts = log(ev)
# ev = np.exp(-tau/ts)
```

```python
def ts_to_ev(lag, ts): 
    return np.exp(-lag/ts)

def ts_to_vamp(lag, ts): 
    ev = ts_to_ev(lag, ts)
    vamp = 1+ev**2
    return vamp
```

```python
ev1 = ts_to_ev(lag, ts1)
ev2 = ts_to_ev(lag, ts2)
```

```python
vamp2_1 = 1 + ev1**2
vamp2_2 = 1 + ev2**2
```

```python
loss1 = 100*(2 - vamp2_1)/2
loss2 = 100*(2 - vamp2_2)/2
loss1, loss2
```
